```python
import pandas as pd
import numpy as np
```

```python
#Reading the train file
train_df1= pd.read_csv("train.csv", parse_dates=['impression_time'] , index_col='impression_id')
# Parsing the dates and indexing the impression ids
train_df1.head()
```

```python
type(train_df1['impression_time'][0]) # Verifying if the dates are converted to timestamp
```

```python
final_train1= train_df1.drop(['app_code'], axis=1)
#Dropping the app_code column and target is_click, does not have an impact
final_train1.head()
```

```python
final_train1 = final_train1[(final_train1['impression_time']>=pd.Timestamp(2018,11,15)) & 
                            (final_train1['impression_time']<pd.Timestamp(2018,12,12))]
```

```python
final_train1 # Trimming the data to the dates between 15 Nov 18 and 11 Dec 18
```

```python
#Reading the log file
train_df2= pd.read_csv("view_log.csv", parse_dates=['server_time'])
# Parsing the dates
train_df2.head()
```

```python
type(train_df2['server_time'][0]) # Verifying if the dates are converted to timestamp
```

```python
final_train2 = train_df2[(train_df2['server_time']>=pd.Timestamp(2018,11,15)) & 
                         (train_df2['server_time']<pd.Timestamp(2018,12,12))]
```

```python
final_train2.index = final_train2.index-1578205 # Reindexing them to 0,1,2...
```

```python
final_train2.head()
```

```python
#Encoding string columns
from sklearn.preprocessing import LabelEncoder
```

```python
os_le = LabelEncoder()
final_train1.loc[:,'os_label'] = os_le.fit_transform(final_train1['os_version'])
device_le = LabelEncoder()
final_train2.loc[:,'device_label'] = device_le.fit_transform(final_train2['device_type'])
final_train1.head()
final_train2.head()
```

```python
final_input1= final_train1.drop(['impression_time','os_version'],axis='columns')
final_input2= final_train2.drop(['server_time','device_type'],axis='columns')
final_input1.reset_index(inplace=True) # Removing impression id index as it might cause confusion
final_input1.drop('impression_id', axis=1, inplace=True)
final_input1.head()
```

```python
#Merging the two dataframe based on user id
#Will be using simple inner join as the binary values cannot be cleaned with random values, cannot extrapolate values
final_input = pd.merge(final_input1,final_input2,on="user_id")
final_input.drop_duplicates(inplace=True)#Clearing the duplicates
final_input.reset_index(inplace=True)#Resetting the index
final_input.drop('index', axis=1, inplace=True)# Again dropping the index column
final_input.head()
train_data = final_input.drop(['is_click','session_id','item_id','device_label'], axis=1)
target = final_input['is_click']
train_data.head()
```

```python
#Setting up a tree and training it
from sklearn import tree
```

```python
model= tree.DecisionTreeClassifier()
model.fit(train_data,target)
```

```python
#Creating test dataframe
#Reading the train file
test_data= pd.read_csv("test.csv", parse_dates=['impression_time'])
# Parsing the dates
test_data.head()
```

```python
test_data.drop(['impression_id', 'impression_time', 'app_code'], axis=1, inplace=True)
test_data.head()
```

```python
osle = LabelEncoder()
test_data.loc[:,'os_label'] = osle.fit_transform(test_data['os_version'])
test_data.head()
```

```python
test_data.drop('os_version', axis=1, inplace=True)
test_data.head()
#train_data.head()
```

```python
train_data.head()
```

```python
answer = model.predict(test_data)
answer # Storing the predicted data in an array
```

```python
#Creating submission file using dataframe
submit_data= pd.read_csv("test.csv")
submit_data.drop(['impression_time','user_id','app_code','os_version','is_4G'], axis=1, inplace=True)
submit_data.to_csv('submission.csv', index=False)
```

```python
#Converting into np array
predict_array = np.array(answer)
np.shape(predict_array)
```

```python
#Writing the predicted array to another file in a column
import csv
with open('prediction.csv', 'a', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    for row in range(0,predict_array.shape[0]):
        myList = []
        myList.append(predict_array[row])
        writer.writerow(myList)
```

```python
# Now bringing them together
a = pd.read_csv("submission.csv")
b = pd.read_csv("prediction.csv", header=None, names=["is_click"])
final_submission = pd.concat([a,b],axis=1)
final_submission.to_csv('final_submission.csv', index=False)

#The End
```

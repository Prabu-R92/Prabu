```
train['InvoiceDate'] = pd.to_datetime(train['InvoiceDate'])
test['InvoiceDate'] = pd.to_datetime(test['InvoiceDate'])
```

```
df = train.append(test, ignore_index = True)
df
```

```
df = df.sort_values('InvoiceDate')
df = df.reset_index(drop = True)
df
```

```
from fastai.tabular import add_datepart

add_datepart(df, 'InvoiceDate', drop=False)
```

```
df_train=df[df['UnitPrice'].isnull()==False].copy()
df_test=df[df['UnitPrice'].isnull()==True].copy()
```

```
#I wonder why quantity is negative! Problem statement does not say it is encoded! So taking the absolute value.
df_train['Quantity'] = np.abs(df_train['Quantity'])
df_test['Quantity'] = np.abs(df_test['Quantity'])
```

```
#Removing outliers!
df_train = df_train[df_train['Quantity'] > 0].reset_index(drop=True)
df_train = df_train[df_train['Quantity'] <= 74000].reset_index(drop=True)

df_train = df_train[df_train['UnitPrice'] > 0.005].reset_index(drop=True)
df_train = df_train[df_train['UnitPrice'] < 38000].reset_index(drop=True)
```

```
from xgboost import XGBRFRegressor
import verstack
from verstack.stratified_continuous_split import scsplit
from sklearn.model_selection import KFold
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
```

```
X,Y=df_train.drop(['InvoiceDate','UnitPrice'],axis=1),df_train['UnitPrice']
Xtest=df_test.drop(['InvoiceDate','UnitPrice'],axis=1)

print(X.shape,Xtest.shape)
X_train,X_val,Y_train,Y_val = scsplit(X,Y,stratify=Y,test_size=0.25,random_state = 2020)
```

```
LG = LGBMRegressor(n_estimators=100, objective='regression', num_leaves=1675, max_depth=11, n_jobs=50, reg_alpha=0.9, reg_lambda=1)
LG.fit(X_train, Y_train)
y_pred = LG.predict(X_val)
print('RMSE:', mean_squared_error(Y_val, y_pred, squared = False))
```

```
LGBMerror=[]
y_pred_LGBM=[]
fold=KFold(n_splits=5,shuffle=True,random_state=2020)
i=1
for train_index, test_index in fold.split(X,Y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model=LGBMRegressor(n_estimators=100, objective='regression', num_leaves=1675, max_depth=11, n_jobs=50, reg_alpha=0.9, reg_lambda=1)
    model.fit(X_train,Y_train,eval_set=[(X_test, Y_test)], early_stopping_rounds=2000, verbose=200)
    preds=model.predict(X_test)[:]
    print("LGBMerror: ",mean_squared_error(Y_test,preds, squared = False))
    LGBMerror.append(mean_squared_error(Y_test,preds, squared = False))
    p = model.predict(Xtest)[:]
    i=i+1
    y_pred_LGBM.append(p)
```

```
XG = XGBRFRegressor(max_depth=12, learning_rate=0.5, n_jobs=50, objective = 'reg:squarederror', random_state=0, reg_alpha=0.9)
XG.fit(X_train, Y_train)
y_pred1 = LG.predict(X_val)
print('RMSE:', mean_squared_error(Y_val, y_pred1, squared = False))
#Same value!! No use!
```

```
XGBRFerror=[]
y_pred_XGBRF=[]
fold=KFold(n_splits=5,shuffle=True,random_state=2020)
i=1
for train_index, test_index in fold.split(X,Y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    model=XGBRFRegressor(max_depth=12, learning_rate=0.5, n_jobs=50, objective = 'reg:squarederror', random_state=0, reg_alpha=0.9)
    model.fit(X_train,Y_train,eval_set=[(X_test, Y_test)], early_stopping_rounds=2000, verbose=200)
    preds=model.predict(X_test)[:]
    print("XGBRFerror: ",mean_squared_error(Y_test,preds, squared = False))
    XGBRFerror.append(mean_squared_error(Y_test,preds, squared = False))
    p = model.predict(Xtest)[:]
    i=i+1
    y_pred_XGBRF.append(p)
```

```
LGBMvalue = np.square(np.mean(y_pred_LGBM,0))
XGBRFvalue = np.square(np.mean(y_pred_XGBRF,0))
submission['UnitPrice'] = LGBMvalue*0.55 + XGBRFvalue*0.45
```
